`train.csv`文件中是训练集，里面是`1-891`号旅客的信息，以及是否获救。  
`test.csv`文件是测试集，里面是`892-1309`号旅客的信息，但是没有`Survived`列，需要我们自己建立模型去预测。  
`gender_submission.csv`是要提交结果的范例文件。

```python
train = pd.read_csv('./input/train.csv')
test = pd.read_csv('./input/test.csv')
```

```python
train.describe()
```

```python
train.info()
```

- PassengerId: 旅客编号  
- Survived: 是否获救，1表示获救，0表示没有获救  
- Pclass: 客舱等级，有1、2、3三个级别 
- Name: 旅客名称  
- Sex: 性别  
- Age: 年龄  
- SibSp:  亲属人数，兄弟姐妹/配偶，即同代的直系亲属  
- Parch:  父母/子女 人数，即不同代的直系亲属  
- Ticket:  船票编号  
- Fare:  票价  
- Cabin:  船舱号  
- Embarked:  登船港口 有3个，C = Cherbourg, Q = Queenstown, S = Southampton

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y = train['Survived']

features = ['Pclass', 'Sex', 'SibSp', 'Parch']

X = pd.get_dummies(train[features])
X_test = pd.get_dummies(test[features])

model = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=2)
model.fit(X, y)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
first_RFC = pd.DataFrame()
```

```python
first_RFC['PassengerId'] = test['PassengerId']
first_RFC['Survived'] = predictions
```

```python
first_RFC.head()
```

```python
first_RFC.to_csv('./output/firt_RFC.csv', index=False)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X, y)
```

```python
y_test = model.predict(X_test)
y_test
```

```python
def one_or_zero(x):
    if x >= 0.5:
        return 1
    else:
        return 0
    
y_pred = np.array([one_or_zero(x) for x in y_test])
```

```python
y_pred
```

```python
LR_result = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': y_pred,
})
```
